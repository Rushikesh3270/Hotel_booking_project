```python
df = df.drop(columns = ['agent', 'company'])
```

```python
df = df.dropna(axis=0)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(9,9))
labels = df['hotel'].value_counts().index
sizes = df['hotel'].value_counts()
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title('Type Hotel Prefered By People');
```

```python
sns.countplot(df['is_canceled']);
```

```python
sns.countplot(data = df, x = 'deposit_type')
plt.title('Graph showing types of deposits', fontsize = 20)
```

```python
lst2 = ['hotel', 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']
period_arrival = df[lst2]
sns.countplot(data = period_arrival, x = 'arrival_date_year', hue = 'hotel')
plt.title('Graph showing number of arrivals per year');
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data = period_arrival, x = 'arrival_date_month', hue = 'hotel', order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December'])
plt.title('Graph showing number of arrivals per month',fontsize=20)
plt.xlabel('Month')
plt.ylabel('Count');
```

```python
plt.figure(figsize=(15,5))

sns.countplot(data = period_arrival, x = 'arrival_date_day_of_month', hue = 'hotel')
plt.title('Graph showing number of arrivals per day', fontsize = 20);
```
